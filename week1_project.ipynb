Feature Engineering

```python
df["hour"] = df["date_time"].dt.hour #9 means 9AM
df["day_of_week"] = df["date_time"].dt.dayofweek #0 indicates monday
df["month"] = df["date_time"].dt.month
```

```python
df[["hour", "day_of_week", "month"]].head()
```

```python
import pandas as pd
x= df[["hour", "day_of_week", "month", "temperature","air_pollution_index","humidity","wind_speed", "wind_direction", "visibility_in_miles", "dew_point", "rain_p_h", "snow_p_h", "clouds_all"]]
y = df["traffic_volume"]
```

```python
x.isnull().sum() #so there is "no null" in the input data.
```

Now training and testing will occur

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42
)

model = LinearRegression()

```

```python
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import mean_absolute_error

lin_MAE = mean_absolute_error(y_test, prediction)
print(lin_MAE)
```

Error is very high since traffic pattern are non linear and LinearRegression struggles with non linearity.

```python
from sklearn.ensemble import RandomForestRegressor

RF_model = RandomForestRegressor(
    n_estimators = 100,
    random_state = 42
)

RF_model.fit(x_train,y_train)
```

```python
RF_predictions = RF_model.predict(x_test)

print(RF_predictions)
```

```python
RF_MAE = mean_absolute_error(y_test, RF_predictions)
print(RF_MAE)
```

```python
avg_traffic = df["traffic_volume"].mean()
error = (RF_MAE/avg_traffic)*100
print("So the error percentage is", error)
```

```python
print("So the mean absolute error in case of Linear regression is", lin_MAE)
print("So the mean absolute error in case of Random Forest Regressor is ", RF_MAE)
```

```python
feature_name = x.columns
importance = RF_model.feature_importances_

importance_df = pd.DataFrame({"Feature": feature_name, "Importance": importance}).sort_values("Importance", ascending = False)
print(importance_df)


```

```python
importance_df["Importance%"] = (importance_df["Importance"]*100).round(2)

print("Feature Importance :")
print("="*45)

print(importance_df[["Feature", "Importance%"]])
```

```python
error = pd.DataFrame({
    "actual" : "y_test", #there are the true value from the dataset.
    "predicted" : "RF_predictions", #this stored the model prediction.
    "error" : abs(RF_predictions - y_test), #absolute difference remove if negative occurs
    "hour" : df.loc[y_test.index, "hour"] #since after train_test_split , rows got shuffled , so corresponding hours required.
}) #created a new table naed "error" which will store actual value, predicted value, prediction mistake and corresponding hour.

hourly_error = error.groupby("hour")["error"].mean().round(2)

print("Hours where model struggled most : ")
print("="*45)
print(hourly_error.sort_values(ascending = False).head())



print("Hours where model struggled least : ")
print("="*45)
print(hourly_error.sort_values().head())


plt.figure(figsize = (12,6))
plt.bar(hourly_error.index, hourly_error.values, color = "purple")
plt.xticks(rotation = 0)

plt.grid(axis ="y", alpha = 0.3)
plt.tight_layout()

plt.xlabel("Hour")
plt.ylabel("Average Error")
plt.title("Average Error by Hour")
plt.show()


```
